```python
# import the Libraries 
import pandas as pd 
```

```python
# load the dataset
path =r'D:\NTI-ML\nti-projects\Medicine-analysis\Medicine.csv'
df = pd.read_csv(path)
```

```python
#Display the first 5 rows of the dataset 
df.head()
```

```python
#Checking the number of missing value in Price and Quantity columns
df.isnull().sum()
```

```python
#fil the missing vlue in the price with the mean 
df['Price'].fillna(df['Price'].mean(), inplace=True)
```

```python
#Display the price column after filling the missing value with the mean
df['Price'].head()
```

```python
#Fill any misssing vlue in the Quantity column with 1
df['Quantity'].fillna(1, inplace=True)
```

```python
# Display the Quantity column after filling the missing value with 1
df['Quantity']
```

```python
#Create a new column called Inventory_Value by inventory_value =price * quantity
df['Inventory_Value'] = df['Price'] * df['Quantity']
```

```python
#Check the data after the changes
df.head()
```

```python
#Create a new DataFrame called well_stocked that contains medicines with a Quantity of 15 or more 
well_stocked = df[df['Quantity'] >= 15]
well_stocked
```

```python
#iloc selection 
#Display row 5 ot 10 and show only the following columns [Medicine , inventory_value]
df.iloc[5:11, [0, 3]]
```

```python
#Reorting 
df[['Inventory_Value']].describe()
```

```python
#Save the cleaned data to a new CSV file named cleaned_medicine_inventory.csv
df.to_csv('cleaned_medicine_inventory.csv', index=False)
```
