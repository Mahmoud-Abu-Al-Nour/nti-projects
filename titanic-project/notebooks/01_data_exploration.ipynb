```python
#show the distribution of the Survived feature in a countplot
sns.countplot(data=df, x="Survived")

plt.show()
```

```python
#Show the distribution of the Age feature in a histogram
sns.histplot(df["Age"], kde=True)

plt.show()
```

```python
#show the distribution of the Fare feature in a histogram
sns.histplot(df["Fare"], kde=True)

plt.show()
```

```python
# show the correlation between the sex and Survived features in a countplot
sns.countplot(data=df,
              x="Sex",
              hue="Survived")

plt.show()
```

```python
#show the correlation between the Pclass and Survived features in a countplot
sns.countplot(data=df,
              x="Pclass",
              hue="Survived")

plt.show()
```

```python
# git the Correlation
numeric_df = df.select_dtypes(include="number")

numeric_df.corr()
```

```python
# show the correlation between the numeric features in a heatmap
plt.figure(figsize=(8,6))

sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap="coolwarm")

plt.show()
```
